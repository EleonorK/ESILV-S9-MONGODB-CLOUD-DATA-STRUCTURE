```python
# CrÃ©er un dictionnaire pour mapper les genres Ã  leurs ID
genre_to_id = dict(zip(df_genre_l['genres_de'], df_genre_l['genres_id']))
```

```python
# Fonction pour convertir les genres en IDs
def genres_to_ids(genres):
    # VÃ©rifier si genres est une chaÃ®ne de caractÃ¨res
    if not isinstance(genres, str):
        return ''  # Retourner une chaÃ®ne vide si ce n'est pas une chaÃ®ne
    return ','.join(str(genre_to_id.get(genre)) for genre in genres.split(',') if genre in genre_to_id)
```

```python
# Apply the function to the 'anime_table_df' DataFrame
df_anime_table['genres_id'] = df_anime_table['genres_de'].apply(genres_to_ids)
```

## ranking

```python
df_merged3 = pd.merge(df_anime_ranking, df_anime[['mal_id', 'title', 'demo_id', 'genres_id', 'studio_id']], on='mal_id', how='left')

df_ranking = df_merged3.dropna()
df_ranking.columns = df_ranking.columns.str.replace('.', '_')
df_ranking = df_ranking.drop_duplicates()
df_ranking
```

```python
def is_multi_genre(genres_id):
    return len(genres_id.split(',')) > 1

# We create a new column 'is_multi_genre' which is True for rows where 'genres_id' contains multiple ids
df_ranking['is_multi_genre'] = df_ranking['genres_id'].apply(is_multi_genre)

# Now we filter the DataFrame to only keep the rows where 'is_multi_genre' is True
df_ranking = df_ranking[df_ranking['is_multi_genre']]

# Drop the 'is_multi_genre' column as it is no longer needed
df_ranking = df_ranking.drop(columns='is_multi_genre')

df_ranking
```
